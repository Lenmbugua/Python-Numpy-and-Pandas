```python
#Pandas Series.str.slice()
# Example #1: In this example, the salary column has been sliced to get values before decimal
# For example, we want to do some mathematical operations and for that we need integer data, so the salary column will be sliced till the 2nd last element(-2 position). 
#Since the salary column is imported as float64 data type, it is first converted to string using the .astype() method.
import pandas as pd
#making DataFrame
data = pd.read_csv('nba.csv')
#removing null values to avoid errors
data.dropna(inplace=True)
#start,stop and step variables
start,stop,step = 0,-2,1
#converting to string data type
data['Salary']=data['Salary'].astype(str)
#slicing till 2nd last element
data['Salary(int)'] = data["Salary"].str.slice(start,stop,step)
#display
data.head(15)
```

```python
#Pandas Series.apply()
#Use Series.apply() function to change the city name to 'Montreal' if the city is 'Rio'.
import pandas as pd
#creating the Series
sr = pd.Series(['New York', 'Chicago','Toronto','Lisbon','Rio'])
#create the index
index_ = ['City1','City2','City3','City4','City5']
#set the index
sr.index = index_
#print the Series
print(sr)
#Now we will use Series.apply() function to change the city name to 'Montreal' if the city is 'Rio'
#change 'Rio' to 'Montreal'
#we have used a lambda function
result = sr.apply(lambda x : 'Montreal' if x == 'Rio' else x)
#print the result
print(result)
```

```python
#Use Series.apply() function to return True if the value in the given series object is greater than 30 else return False.
import pandas as pd
#creating the Series
sr = pd.Series([11,21,8,18,65,18,32,10,5,32,None])
#Create the index
#apply yearly frequency
index_ = pd.date_range('2010-10-09 08:45',periods = 11, freq = 'Y')
#set the index
sr.index = index_
#print the Series
print(sr)
#Now we will use Series.apply() function to return True if a value in the given series object is greater than 30 else return False.
#return true if greater than 30
#else return False
result = sr.apply(lambda x : True if x > 30 else False)
#print the result
print(result)
```

```python
#Applying User-Defined Function to Every Row of Pandas DataFrame
import pandas as pd

# Function to add values from columns A, B, and C
def add_values(row):
    return row['A'] + row['B'] + row['C']

def main():
    # Create a dictionary with three fields
    data = {
        'A': [1, 2, 3],
        'B': [4, 5, 6],
        'C': [7, 8, 9]
    }

    # Convert the dictionary into a DataFrame
    df = pd.DataFrame(data)
    print("Original DataFrame:\n", df)

    # Apply the user-defined function to every row
    df['sum'] = df.apply(add_values, axis=1)

    # Print the new DataFrame
    print('\nAfter applying Function:')
    display(df)

# Ensure main() runs only if this script is executed directly
if __name__ == '__main__':
    main()
```

```python
#Apply Lambda to Every Row of DataFrame
import pandas as pd
#Function to add
def add(a,b,c):
    return a + b + c

def main():
    #create a dictionary with three fields each
    data = {
        'A': [1,2,3],
        'B': [4,5,6],
        'C': [7,8,9]
    }
    #convert the dictionary into a DataFrame
    df = pd.DataFrame(data)
    print('Original DataFrame:\n',df)
    df['sum'] = df.apply(lambda row: add(row['A'], row['B'], row['C']), axis=1)
    print('\nAfter Applying Function: ')
    #displaying the new DataFrame
    display(df)

if __name__ == '__main__':
    main()
```

```python
#Apply NumPy.sum() to Every Row
import pandas as pd
import numpy as np
def main():
    #create a dictionary with five fields each
    data = {
        'A': [1,2,3],
        'B': [4,5,6],
        'C': [7,8,9]
    }
    #convert the dictionary into a DataFrame
    df = pd.DataFrame(data)
    print("Original DataFrame:\n",df)
    #applying function to each row in the DataFrame and storing result in a new column
    df['Sum'] = df.apply(np.sum, axis=1)
    print('\nAfter Applying Function: ')
    #printing the new DataFrame
    display(df)

if __name__ == '__main__':
    main()
```

```python
#Normalizing DataFrame Column Values Using Custom Function in Pandas
import pandas as pd
def normalize(x, y):
    x_new = ((x - np.mean([x, y])) /
             (max(x, y) - min(x, y)))
    #print(x_new)
    return x_new

def main():
    #create a dictionary with three fields each
    data = {
        'X': [1,2,3],
        'Y': [45,65,89]
    }
    #convert the dictionary into a DataFrame
    df = pd.DataFrame(data)
    print("Original DataFrame:\n",df)
    df['X'] = df.apply(lambda row: normalize(row['X'], row['Y']), axis=1)
    print('\nNormalized:')
    display(df)

if __name__ == '__main__':
    main()
```

```python
#Applying Range Generation Function to DataFrame Rows in Pandas
import pandas as pd

# Function to generate range
def generate_range(n):
    n = int(n)
    lower_limit = n // 10 * 10
    upper_limit = lower_limit + 10
    return f"{lower_limit}-{upper_limit}"

# Function to replace each item in the row with its range
def replace(row):
    for i, item in enumerate(row):
        row.iloc[i] = generate_range(item)
    return row

def main():
    # Original numeric data
    data = {
        'A': [0, 2, 3],
        'B': [4, 15, 6],
        'C': [47, 8, 19]
    }

    # Convert dictionary to DataFrame
    df = pd.DataFrame(data)

    print('Before applying function:')
    print(df)

    # Convert entire DataFrame to object type (to allow string assignment)
    df = df.astype(str)

    # Apply the function row-wise
    df = df.apply(lambda row: replace(row), axis=1)

    print('\nAfter Applying Function:')
    display(df)

if __name__ == '__main__':
    main()
```

```python
#Adding Rows with Missing Columns
#Handling scenarios where the new row has fewer columns than the original DataFrame.
import pandas as pd
data = {"Name": ["Alice", "Bob"], "Age": [25, 30], "City": ["NY", "LA"]}
df = pd.DataFrame(data)

# Add a row missing one column
new_row = {"Name": "Charlie", "Age": 35}  # No "City"
df = pd.concat([df, pd.DataFrame([new_row])], ignore_index=True)
df
```

```python
#Pandas GroupBy
#Step 1: Splitting Data into Groups
#Group data by a single key: In order to group data with one key we pass only one key as an argument in groupby function. Here we will group the data by the Name column.
import pandas as pd
data1 = {'Name':['Jai', 'Anuj', 'Jai', 'Princi',
                 'Gaurav', 'Anuj', 'Princi', 'Abhi'],
        'Age':[27, 24, 22, 32,
               33, 36, 27, 32],
        'Address':['Nagpur', 'Kanpur', 'Allahabad', 'Kannuaj',
                   'Jaunpur', 'Kanpur', 'Allahabad', 'Aligarh'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd',
                         'B.Tech', 'B.com', 'Msc', 'MA']}
df = pd.DataFrame(data1)
df
grouped = df.groupby('Name')
grouped.groups
#This prints a dictionary showing the index positions of the rows belonging to each group.
gk = df.groupby('Name')
gk.first()
```

```python
#Grouping data with multiple keys : In order to group data with multiple keys, we pass multiple keys in groupby function. Here we group a data of "Name" and "Qualification" together using multiple keys in groupby function. 
import pandas as pd
df.groupby(['Name','Qualification'])
df.groupby(['Name','Qualification']).groups
```
